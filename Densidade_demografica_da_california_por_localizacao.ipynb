## Objetivo
Faremos análises dos dados da população das cidades da California, medindo proporções de área e densidade demográfica

```python

import pandas as pd
cities = pd.read_csv("california_cities.csv")
print(cities.head())
```

```python
# extraindo os dados que são realmente relevantes
latitude, longitude = cities["latd"], cities["longd"]
populacao, area = cities["population_total"], cities["area_total_km2"]
```

```python
import numpy as np #biblioteca para operações multidimensionais
import matplotlib.pyplot as plt #para gráficos
import seaborn #para fins de personalização dos gráficos (mudamos as cores)

#espalhando os pontos adicionando cores, mas sem rótulos
seaborn.set()
plt.scatter(longitude, latitude, label=None, c=np.log10(populacao),
            cmap='viridis', s=area, linewidth=0, alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('Longitude')
plt.ylabel('Longitude')
plt.colorbar(label='log$_{10}$(população)')
plt.clim(3, 7)
# Agora a criação da legenda, vamos preencher listas vazias com rótulos e tamanhos desejados
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.title("Área e população das cidades de California")
plt.show()
```

Concluimos que a maior densidade demográfica está localizada nas cidades de mais baixa latitude
